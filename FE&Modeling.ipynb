```python
def reduce_mem_usage(df, verbose=True):
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    start_mem = df.memory_usage().sum() / 1024**2    
    for col in df.columns:
        col_type = df[col].dtypes
        if col_type in numerics:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)  
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)    
    end_mem = df.memory_usage().sum() / 1024**2
    if verbose: print('Mem. usage decreased to {:5.2f} Mb ({:.1f}% reduction)'.format(end_mem, 100 * (start_mem - end_mem) / start_mem))
    return df
```

```python
class GAN():
    def __init__(self):

        self.non_seq_shape=(64,)
        optimizer_SGD= SGD()
        optimizer_adam= Adam()
        
        # Build and compile the discriminator
        self.discriminator = self.build_discriminator()
        self.discriminator.compile(loss='binary_crossentropy',optimizer=optimizer_SGD, metrics=['accuracy'])
        
        # Build the generator
        self.generator1 = self.build_generator1()
        noise_non_sequential=Input(shape=(64,))
       
        non_sequential_fake=self.generator1(noise_non_sequential)
        
        # combined generator and discriminator for generator training
        self.discriminator.trainable = False  #discriminator is not trainable when training generator
        validity = self.discriminator(non_sequential_fake)
        self.combined = Model(noise_non_sequential, validity)
        self.combined.compile(loss='binary_crossentropy', optimizer=optimizer_adam)
        
            

    def build_discriminator(self):
        in_layers = list()
        em_layers = list()
        for i in range(len(categorical_f)):
            # calculate the number of unique inputs
            labels = set(list(train[categorical_f[i]].unique())+list(test[categorical_f[i]].unique()))
            n_labels=len(labels)
            # define input layer
            in_layer = Input(shape=(1,))
            # define embedding layer
            n=np.log(n_labels)//np.log(2)+1
            n=int(n)
            em_layer = Embedding(n_labels, n,input_length=1)(in_layer)
            #em_layer=Flatten()(em_layer)
            # store layers
            in_layers.append(in_layer)
            em_layers.append(em_layer)

        input_ctn=Input(shape=(len(ctn_f),),name="dense_input")
        output_ctn=Dense(128)(input_ctn)
        in_layers.append(input_ctn)
        
       
        merge = concatenate(em_layers)
        output_cat=Flatten()(merge)
        merge = concatenate([output_cat,output_ctn])
        m = Dense(512, activation='relu')(merge)
        m =BatchNormalization()(m)
        m =Dropout(0.5)(m)
        m =Dense(256, activation='relu')(m)
        m =BatchNormalization()(m)
        m =Dropout(0.5)(m)
        m =Dense(128, activation='relu')(m)
        m =BatchNormalization()(m)
        m =Dropout(0.5)(m)
        m =Dense(64, activation='relu')(m)
        m =BatchNormalization()(m)
        m =Dropout(0.5)(m)
        m =Dense(32, activation='relu')(m)
        m =BatchNormalization()(m)
        m =Dropout(0.5)(m)
        output = Dense(1, activation='sigmoid')(m)


        model = Model(inputs=in_layers, outputs=output)
        return model
    
    def build_generator1(self):
        inputs=Input(shape=(64,))

        g=Dense(32)(inputs)
        g=LeakyReLU(alpha=0.2)(g)
        g=BatchNormalization()(inputs)
        g=Dense(32,activation='tanh')(g)
       # g=LeakyReLU(alpha=0.2)(g)

        out_layers=[]
        for i in range(len(categorical_f)):
            #out = Dense(2)(g)
            out=LeakyReLU(alpha=0.2)(g)
            out = Dense(1,activation='tanh')(out)
            out_layers.append(out)
        out=Dense(128)(g)
        out=LeakyReLU(alpha=0.2)(out)
        out=Dense(len(ctn_f),activation='tanh')(g)
        out_layers.append(out)


        model = Model(inputs=inputs, outputs=out_layers)

        return model
    
    

    def train(self,x_train,x_val,y_train,y_val,epochs=200, batch_size=128):
        auc_list=[]
        ks_list=[]
        auc_progress = []
        half_batch = int(batch_size / 2)

        noise_until = epochs


        for epoch in range(epochs):
            # Select a random half batch of real benign users data
            idx = np.random.randint(0, y_train.shape[0], half_batch)
    
            non_sequential=[i[idx] for i in x_train]

           # Sample noise and generate a half batch of new fake data
            noise1 = np.random.normal(0, 1, (half_batch, 64))
               
            non_sequential_fake=self.generator1.predict(noise1)
            #non_sequential_fake=[non_sequential_fake_raw[i] for i in range(44)]
            #non_sequential_fake.append(non_sequential_fake_raw[44])
            #print(non_sequential_fake)

            valid = np.ones((half_batch, 1))
            fake = np.zeros((half_batch, 1))

            # Train the discriminator
            
            d_loss_real = self.discriminator.train_on_batch(non_sequential, valid)
            d_loss_fake = self.discriminator.train_on_batch(non_sequential_fake, fake)

 
            d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)



            #  Train Generator
  
            noise1 = np.random.normal(0, 1, (batch_size, 64))
                
            validity = np.ones((batch_size, 1))

            g_loss = self.combined.train_on_batch(noise1, validity)
           

            if epoch % 5 == 0:
                print ("%d [D loss: %f, acc: %.2f%%] [G loss: %f]" % (epoch, d_loss[0], 100*d_loss[1], g_loss))

            if epoch % 50 == 0:
                y_pred = self.discriminator.predict(x_val,batch_size=batch_size)
                (fpr, tpr, thresholds) = roc_curve(y_val,y_pred)
                area = auc(fpr,tpr)
                auc_list.append(area)

                ks=(tpr-fpr)
                max_ks=np.max(ks)
                ks_list.append(max_ks)
                print('Epoch: {}, auc: {:.5f}, ks: {}'.format(epoch,area,max_ks))

        return auc_list,ks_list
```
